# __OVERALL TOSS DETAILS__

```python
df1_toss=df1[["team1","team2","toss_winner","toss_decision","winner"]]
display(df1_toss)
```

### __DETAILS OF EVERY TEAM WHO WON THE TOSS AND ELECTED TO FILED FIRST__

```python
df1_toss1=df1_toss[df1_toss["toss_decision"]=="field"][["toss_winner","toss_decision","winner"]]
df1_toss1
```

```python
#finding the count of team which won toss and elected to filed and arranging them in descinding order
toss_field_count=df1_toss1.groupby('toss_winner')[['winner']].count().sort_values(by=['winner'], ascending=False)
```

```python
toss_field_count
```

__Kolkata Knight Riders	 won the toss 86 times and elected to filed first__

```python
plt.bar(toss_field_count.index,toss_field_count.winner,color='#33a1c9')
plt.xticks(fontsize=18,rotation =90)
plt.yticks(fontsize=20)
plt.title("TEAMS WON THE TOSS AND ELECTED TO FIELD", fontsize=20)
plt.rcParams['figure.figsize']=[15,5];
```

### __DETAILS OF EVERY TEAM WHO WON THE TOSS AND ELECTED TO BAT FIRST__

```python
df1_toss2=df1_toss[df1_toss["toss_decision"]=="bat"][["toss_winner","toss_decision","winner"]]
df1_toss2
```

```python
##finding the count of team which won toss and elected to filed and arranging them in descinding order
toss_bat_count=df1_toss2.groupby('toss_winner')[['winner']].count().sort_values(by=['winner'], ascending=False)
toss_bat_count
```

```python
plt.bar(toss_bat_count.index,toss_bat_count.winner,color='#cdaa7d')
plt.xticks(fontsize=20,rotation =90)
plt.yticks(fontsize=20)
plt.title("TEAMS WON THE TOSS AND ELECTED TO BAT", fontsize=20)
plt.rcParams['figure.figsize']=[20,8];
```

```python
#merging both tables
final= toss_bat_count.merge(toss_field_count, how='left',on='toss_winner')
```

```python
final.plot(kind='bar')
plt.xticks(fontsize=20)
plt.title(' TOSS:win and choose', fontsize=25)
plt.legend(['bat','field'], fontsize=20)
plt.rcParams['figure.figsize']=[20,8];
```

# UMPIRE

```python
df1_um=df1[["match_id","umpire1","umpire2"]]
df1_um
```

```python
df1_um1=df1_um.groupby("umpire1").count()

df1_um1
```

```python
df1_um2=df1_um.groupby("umpire2").count()
df1_um2
```

```python
ump=pd.merge(df1_um2,df1_um1, left_index=True,right_index=True)
```

```python
ump[['umpire1','umpire2']] 
ump["total"]=ump.umpire1+ump.umpire2
```

```python
ump_count=ump['total'].sort_values(ascending=False)
ump_count
```

__Above are the names of umpires and the total number of matches they umpired__

```python
ump_count.head(10).plot(kind='bar',color='#00c78c' )
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('matches',fontsize=20)
plt.title('UMPIRES AND MATCHES', fontsize=25);
```
